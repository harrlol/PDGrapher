```python
import torch
import pandas as pd
```

```python
##Chemical perturbations
cell_line = ['A375', 'A549', 'BT20', 'HA1E', 'HELA', 'HT29', 'MCF7', 'MDAMB231', 'PC3', 'VCAP']

for cl in cell_line:
    split = torch.load('processed/splits/chemical/{}/random/5fold/splits.pt'.format(cl))

    print('Cell line: {}'.format(cl))

    #Loads data
    b_data = torch.load('processed/torch_data/chemical/real_lognorm/data_backward_{}.pt'.format(cl))
    f_data = torch.load('processed/torch_data/chemical/real_lognorm/data_forward_{}.pt'.format(cl))

    for i in range(1, 6):
        print(' Split: {}'.format(i))

        error = False
        #Backward data
        d = b_data
        if len(d) != len(split[i]['train_index_backward']) + len(split[i]['val_index_backward']) + len(split[i]['test_index_backward']):
            print('length is not the same for splits: {}'.format(cl))
            error = True
        
        #Forward data
        d = f_data
        if len(d) == 0:
            if split[i]['train_index_forward'] is not None or split[i]['val_index_forward'] is not None or split[i]['test_index_forward'] is not None:
                print('split is not None when it should be (no forward data)')
                error = True
        else:
            if len(d) != len(split[i]['train_index_forward']) + len(split[i]['val_index_forward']) + len(split[i]['test_index_forward']):
                print('length is not the same for splits: {}'.format(cl))
                error = True
        
        if not error:
            print(' Passed test')
        else:
            print(' Failed test')
    



```
